```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
```

```python
# Encode categorical features
df['Sex'] = df['Sex'].astype('category').cat.codes
df['BP'] = df['BP'].astype('category').cat.codes
df['Cholesterol'] = df['Cholesterol'].astype('category').cat.codes
df['Drug'] = df['Drug'].astype('category').cat.codes
```

```python
X = df.drop('Drug', axis=1)
y = df['Drug']
```

```python
print(X.shape)
print(y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.57, random_state=42)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy*100}')
```

```python
# Classification report
print(classification_report(y_test, y_pred))
```

```python
print(confusion_matrix(y_test, y_pred))
```
